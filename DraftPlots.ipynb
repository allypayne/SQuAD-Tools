```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import detections as data
```

```python
data_set= pd.read_csv('galfit_spectral_refined_table_for_kirill_w_bval_3.csv')
```

## First read in the detections & non detections

#### (you can adjust the ion from H_I or O_VI and change the virial radius, but typically we use 1)
#### I am including the basic detections here as well but we will just use the safe ones

```python
detections_h1_1rad= data.sort_detections(data_set, ion= 'H_I', vir_rad= 1)
nondetections_h1_1rad= data.safe_non_detections(data_set, ion='H_I', vir_rad=1)
nondetections_h1_1rad_basic= data.basic_non_detections(data_set, ion='H_I', vir_rad=1)
```

## Now use "make_plot" to display the data

#### Refer to the detections.py code to see all of the available parameters

#### We always use the colormap "Reds" for hydrogen and "Blues" for oxygen
#### set_xmax and y_max should be consistent with the virial radius value
#### Adjust the ion and the color_bar to see different plots (they should always be consistent with eachother)

## Here we are looking at all column density data for galaxies that meet our selection criteria with a detection within one virial radius of the galaxy for H I

```python
data.make_plot(data_set, detections_h1_1rad, nondetections_h1_1rad, ion='H_I', color_bar='H_I_total_log10_N', cmap_color='Reds', set_xmax=1, set_ymax=1, set_title='1 Vir Rad H I Column Density')
```

## With a few simple adjustments we can do the same for our oxygen data

```python
detections_o6_1rad= data.sort_detections(data_set, ion= 'O_VI', vir_rad= 1)
nondetections_o6_1rad= data.safe_non_detections(data_set, ion='O_VI', vir_rad=1)
nondetections_o6_1rad_basic= data.basic_non_detections(data_set, ion='O_VI', vir_rad=1)
data.make_plot(data_set, detections_o6_1rad, nondetections_o6_1rad, ion='O_VI', color_bar='O_VI_total_log10_N', cmap_color='Blues', set_xmax=1, set_ymax=1, set_title='1 Vir Rad O VI Column Density')
```
